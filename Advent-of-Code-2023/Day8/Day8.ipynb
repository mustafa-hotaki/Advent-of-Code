```python
data = ''
with open("input.txt", "r") as f:
    for line in f:
        data += (line)

print('First 10 data points:', data[:10])
print('Number of data points:', len(data))
```

```python
instructions, mappings_raw = data.split('\n\n')

mappings = {}

for line in mappings_raw.split('\n'):
    key, val = line.split(' = ')
    mappings[key] = tuple(val[1:-1].split(', '))

mappings
```

```python
def advance_step(node, inst):
    next = mappings[node]
    pointer = None
    if inst == 'L':
        pointer = next[0]
    else:
        pointer = next[1]

    return pointer

advance_step('AAA', 'R')
```

```python
counter = 0
pointer = 'AAA'
while True:
    if pointer == 'ZZZ':
        break
    inst = instructions[counter % len(instructions)]
    pointer = advance_step(pointer, inst)
    counter += 1

counter
```

```python
counter = 0

pointers = [k for k in mappings.keys() if k.endswith('A')]

patterns = [[] for p in pointers]

limit = 5
while True:
    inst = instructions[counter % len(instructions)]

    for i, p in enumerate(pointers):
        pointers[i] = advance_step(p, inst)
        if p.endswith('Z'):
            if len(patterns[i]) <= limit:
                patterns[i] += [counter]

    counter += 1

    if sum([len(x) > limit for x in patterns]) == len(pointers):
        break

print(*patterns, sep='\n')
```
